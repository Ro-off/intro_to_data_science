Стовбець **`published`** (час публікації) містить рядки. Щоб ми мали змогу працювати з цими даними як з датою/часом публікації, приведемо їх до типу `datetime`:

```python
print(df.published.dtype)
df['published'] = pd.to_datetime(df.published, yearfirst = True)
print(df.published.dtype)
```

Створимо декілька стовбців на основі даних про час публікації:

```python
df['year'] = [d.year for d in df.published]
df['month'] = [d.month for d in df.published]

df['dayofweek'] = [d.isoweekday() for d in df.published]
df['hour'] = [d.hour for d in df.published]
```

## 1\. В якому місяці (і якого року) було більше всього публікацій?

* березень 2016
* **березень 2015**
* квітень 2015

```python
# Групуємо за роком і місяцем, потім рахуємо кількість публікацій
monthly_publications = df.groupby(['year', 'month']).size().reset_index(name='count')

# Знаходимо місяць і рік з найбільшою кількістю публікацій
max_publications = monthly_publications.loc[monthly_publications['count'].idxmax()]

# Візуалізуємо кількість публікацій за місяцями
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_publications, x='month', y='count', hue='year')
plt.title('Кількість публікацій за місяцями та роками')
plt.xlabel('Місяць')
plt.ylabel('Кількість публікацій')
plt.legend(title='Рік')
plt.show()

max_publications
```

## 3\. Коли найкраще всього опублікувати статтю?

* Більш всього переглядів набирають статті, опубліковані в 12 годин дня
* В опублікованих о 10-й ранку постів більше всього коментарів
* **Більше всього переглядів набирають статтті, опубліковані в 6 годин ранку**
* Максимальне число коментарів на гіктаймсі набрала стаття, опублікована в 9 годин вечора
* На хабрі денні статті комментують частіше, ніж вечірні

```python
# Групуємо за годиною і обчислюємо середню кількість переглядів і коментарів
hourly_stats = df.groupby('hour').agg({'views': 'mean', 'comments': 'mean'}).reset_index()

# Побудова графіку середньої кількості переглядів за годинами
plt.figure(figsize=(14, 7))
sns.lineplot(data=hourly_stats, x='hour', y='views', marker='o')
plt.title('Середня кількість переглядів за годинами публікації')
plt.xlabel('Година публікації')
plt.ylabel('Середня кількість переглядів')
plt.grid(True)
plt.show()

# Побудова графіку середньої кількості коментарів за годинами
plt.figure(figsize=(14, 7))
sns.lineplot(data=hourly_stats, x='hour', y='comments', marker='o')
plt.title('Середня кількість коментарів за годинами публікації')
plt.xlabel('Година публікації')
plt.ylabel('Середня кількість коментарів')
plt.grid(True)
plt.show()
```

## 5\. Порівняйте суботи і понеділки

Чи правда, що по суботам автори пишуть в основному вдень, а по понедількам — в основному вечером?

```python
# Filter data for Saturdays and Mondays
saturdays = df[df['dayofweek'] == 6]
mondays = df[df['dayofweek'] == 1]

# Plot the distribution of publication hours for Saturdays and Mondays
plt.figure(figsize=(14, 7))
sns.histplot(saturdays['hour'], bins=24, kde=True, color='blue', label='Saturday', stat='density')
sns.histplot(mondays['hour'], bins=24, kde=True, color='red', label='Monday', stat='density')
plt.title('Distribution of Publication Hours on Saturdays and Mondays')
plt.xlabel('Hour of the Day')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
